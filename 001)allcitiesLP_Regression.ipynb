```python
X=df[['City_1', 'City_2', 'City_3', 'City_5', 'City_6', 'City_7',
       'City_8', 'City_9', 'City_10', 'City_11', 'City_12', 'City_13',
       'City_14', 'City_15', 'City_16', 'City_17', 'City_18', 'City_20',
       'City_21', 'City_22', 'City_24', 'City_25', 'City_26', 'City_27',
       'City_29', 'City_30', 'City_31', 'City_32', 'City_33', 'City_34']]
```

```python
y=df[['City_35']]  #prediction pm2.5
```

```python
X
```

```python
### spliting dataset into training and testing part
from sklearn.model_selection import train_test_split
```

```python
### splitting dataset into training and testing part(50% training and 50% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state =12)
```

```python
### Importing LazyRegressor
from lazypredict.Supervised import LazyRegressor
### Importing dataset available in sklearn
from sklearn import datasets
from sklearn.utils import shuffle
import numpy as np
```

```python
### fitting data in LazyRegressor because here we are solving Regression use case. 
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
```

```python
### fitting data in LazyClassifier
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
### lets check which model did better on Air Pollution Dataset
print(models)
```
